# Рассмотрим основные метрики и измерения

```python
conversion_rate = df['purchase_made'].mean() * 100  #посчитаем конверсию, чтобы определить насколько наш сайт эффективен в продажах, то есть какой процент заходивших совершает покупку
print(f"Конверсия: {round(conversion_rate,2)}%") #посчитаем через функцию .mean() по поиску среднего значения по столбцу и умножим на 100%
```

```python
average_time_on_site = df['time_on_site'].mean()   #посчитаем среднее время на сайте
print(f"Среднее время на сайте: {round(average_time_on_site,2)} секунд")  #посчитаем через функцию .mean() по поиску среднего значения по столбцу и умножим на 100%
```

```python
average_clicks_per_user = df['clicks'].mean()   #посчитаем среднее число кликов
print(f"Среднее количество кликов на пользователя: {round(average_clicks_per_user,2)}")  #посчитаем через функцию .mean() по поиску среднего значения по столбцу и умножим на 100%
```

```python
bounce_rate = (df['page_views'] == 1).mean() * 100 #посчитаем сколько людей после посещения 1 странички отказывлись от покупки и выходили
print(f"Показатель отказов: {round(bounce_rate,2)}%") #посчитаем через функцию .mean() по поиску среднего значения по столбцу и умножим на 100%
```

```python
average_pages_per_user = df['page_views'].mean() #посчитаем среднее число страниц на пользователя
print(f"Среднее количество страниц на пользователя: {round(average_pages_per_user,2)}")#посчитаем через функцию .mean() по поиску среднего значения по столбцу и умножим на 100%
```

```python
first_page_purchase = df[(df['page_views'] == 1)&(df['purchase_made']==1)].shape[0]/10000*100 #посчитаем сколько людей совершили покупку после посещения 1 страницы
print(f"Доля людей, совершивших покупку с первой страницы: {round(first_page_purchase,2)}%") #посчитаем через фильтрацию таблицы: доля строк, подходящих по условию, на общее число 
```

```python
users = df['user_id'].unique().shape[0] #посмотрим сколько у нас всего пользователей
print(f'Всего пользователей: {users}') #посчитаем через уникальные значения в столбце 'user_id'
```

```python
av_user_visit = df['user_id'].value_counts().mean() #посмотрим среднее число посещений на пользователя
print(f'Среднее число посещений сайта одним пользователем: {round(av_user_visit)}') #узнаем это через функцию .mean() и .value_counts(), считающее число строк по уникальному значению столбца
```

Сравним показатели для купивших и не купивших, чтобы понять где покупатели теряются в нашей воронке покупок

```python
# Проанализируем воронку покупок, сгруппировав ее на купивших и не купивших и посчитав для них средние показатели по времени на сайте, числу кликов и страниц
purchase_funnel = df.groupby('purchase_made').agg({
    'user_id': 'count',
    'page_views': 'mean',
    'clicks': 'mean',
    'time_on_site': 'mean',
}).reset_index()
```

```python
print("\nАнализ воронки покупок:")
print(purchase_funnel)
```

Для более показательного анализа нам бы понадобились данные по каждой страницу сайт, чтобы провести аналитику по этим метрикам по ним и определить, где происходят задержки и выходы. Также, нам бы очень помогла тепловая карта по кликам

А теперь мы можем использвать время на сайте и количество кликов для подсчета корреляции Пирсона каждого из этих показателей с фактом покупки. Исходя из этого мы сможем найти проблемные зоны и дать рекомендации

```python
correlation_time = df['time_on_site'].corr(df['purchase_made']) #посчитаем коррелляцию Пирсона для факта покупки и времени на сайте через специальную функцию библиотеки Пандас 
print(f'Корреляция между фактом покупки и времени на сайте: {round(correlation_time, 4)}')
```

```python
correlation_clicks = df['clicks'].corr(df['purchase_made']) #посчитаем коррелляцию Пирсона для факта покупки и количества кликов через специальную функцию библиотеки Пандас 
print(f'Корреляция между фактом покупки и количеством кликов:{round(correlation_clicks, 4)}')
```

```python
correlation_pages = df['page_views'].corr(df['purchase_made']) #посчитаем коррелляцию Пирсона для факта покупки и количества страниц через специальную функцию библиотеки Пандас 
print(f'Корреляция между фактом покупки и количеством страниц:{round(correlation_pages, 4)}')
```

Из корреляций видно, что чем больше времени пользователь проводит на сайте, тем выше вероятность, что он совершит покупку. Для количества кликов закономаерность обратная.

Но нам также видно, что значение корреляций незначительное. Для этого мы посчитаем статистику критерия корреляции Пирсона и построим для нее доверительный интервал, принимая за нулеву гипотезу, что время/число кликов и факт покупки явдяются независимыми показателями. 

Построим доверительный интервал для наших корреляций с уровень доверия 0.95 для двустороння альтернативы. Тогда уровень значимости 0.25. 

```python
import scipy #импортируем библиотеку для работы с распределением Стъюдента
trust = scipy.stats.t.interval(confidence = 0.25, df = 9998, loc = 0, scale=1) #с помощью функции из этой библиотеки найдем доверительный интервал с вышеперечисленными параметрами
print(f'Доверительный интервал: {trust}')
```

```python
# Теперь посчитаем наблюдаемый показатель (t наблюденное) для наших корреляций, преобразовав по формуле корреляцию Спирмена для распределения Стъюдента
t_watched_time = (correlation_time/(1-correlation_time)**0.5)*(10000-2)**0.5
print(f'Наблюдаемый показатель для корреляции времени на сайте: {t_watched_time}')
t_watched_clicks = (correlation_clicks/(1-correlation_clicks)**0.5)*(10000-2)**0.5
print(f'Наблюдаемый показатель для корреляции числа кликов: {t_watched_clicks}')
t_watched_pages = (correlation_pages/(1-correlation_pages)**0.5)*(10000-2)**0.5
print(f'Наблюдаемый показатель для корреляции числа страниц: {t_watched_pages}')
```

Мы видим, что корреляция по времени на сайте/числу страниц входит в доверительный интервал. Следовательно, нулевую гипотезу мы не отвергаем и принимаем показатели совершения покупки и времени на сайте/числу страниц как независимые переменные.

Корреляция по кликам не входит в доверительный интервал. Следовательно, нулевую гипотезу мы отвергаем и решаем, что справедливо сказать, что, чем больше кликов совершает пользователь, тем меньше вероятность покупки.
